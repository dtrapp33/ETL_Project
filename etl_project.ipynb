```python
import pandas as pd
#!pip install forex_python
from forex_python.converter import CurrencyRates
```

```python
boat_data = "boat_data.csv"

# Read the boat data
boat_df = pd.read_csv(boat_data)

# Display the data table for preview
boat_df.head(5)
```

```python
data_df = boat_df[['Price', 'Boat Type', 'Manufacturer', 'Type', 'Year Built', 'Length', 'Width', 'Material', 
                                        'Location', 'Number of views last 7 days']]
data_df
```

```python
#Create column header for Currency from Price column
data_df['Currency'] = data_df['Price'].str[:3]
data_df.head()
```

```python
#Remove Currency prefix from Price column
data_df['Price'] = data_df['Price'].str.slice(3)
data_df.head()
```

```python
#Reorder the dataframe so that Currency is listed first
reorder_df = data_df[['Currency', 'Price', 'Boat Type', 'Manufacturer','Type','Year Built','Length','Width','Material',
                      'Location','Number of views last 7 days']]
reorder_df.head()
```

```python
#Check the currencies listed.
#Note that Â£ might be considered GBP 
reorder_df['Currency'].value_counts()
```

```python
reorder_df.Price = reorder_df.Price.astype(float)
(print(reorder_df.dtypes))
```

```python
#Use a date to reference the currency conversation rate
reorder_df['date'] = pd.Timestamp('2023-09-05')
reorder_df.head()
```

```python
#Currency conversion dataframe
cc = pd.DataFrame(
    {'from':['EUR','CHF','EUR','DKK','EUR','Â£'],'to':['CHF','EUR','DKK','EUR','Â£','EUR'],'rate':[.95,1.05,7.45,.13,1.00,1.00]},
    

)
cc
```

```python
#Applying conversion to EUR:
reorder_df['price in EUR'] = reorder_df['Price']*reorder_df.merge(cc[cc.to=='EUR'].append({'from':'EUR','to':'EUR','rate':1},ignore_index=True),left_on='Currency',right_on='from',how='left')['rate']
reorder_df
```

```python
#Applying conversion to USD:
reorder_df['price in USD'] = round(reorder_df['price in EUR']*1.07,2)
reorder_df.head()
```

```python
reorder_df['Location'] = reorder_df['Location'].str.encode('latin-1').str.decode('utf-8')
reorder_df.head()
```

```python
reorder_df['Manufacturer'] = reorder_df['Manufacturer'].str.encode('latin-1').str.decode('utf-8')
reorder_df
```

```python
reorder_df['Manufacturer'].value_counts()
```

```python
reorder_df['Location'] = reorder_df['Location'].str.replace('»',',')
reorder_df.head()
```

```python
columnsplit = reorder_df['Location'].str.split(" , ", expand=True)
final_df = reorder_df.assign(Country=columnsplit[0],City=columnsplit[1])
final_df.head()
```

```python
typesplit = final_df['Type'].str.split(",", expand=True)
final_df_1 = final_df.assign(Condition=typesplit[0],Fuel=typesplit[1])
final_df_1.head()
```

```python
final_df_1 = final_df_1.drop(['Type','Location'], axis=1)
final_df_1.head()
```

```python
final_df_1.columns.values
```

```python
ETL_Final = final_df_1[['Currency', 'Price', 'Boat Type', 'Manufacturer', 'Year Built',
       'Length', 'Width', 'Material', 'Number of views last 7 days',
       'date', 'price in EUR', 'price in USD', 'City','Country',
       'Condition', 'Fuel']]
ETL_Final.head()
```

```python
ETL_Final = ETL_Final.rename(columns={'date': 'Date', 'price in EUR': 'Price in EUR', 'price in USD': 'Price in USD',
                                      'Length': 'Length (meters)','Width': 'Width (meters)'})
ETL_Final
```
